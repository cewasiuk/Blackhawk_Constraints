```python
#Blackbody LT Computation
import math as ma 
M_sun = 1.989e30
M = 1*M_sun #initial mass in kg
M_rem = 0 #Remnant mass in kg
a = 0

#Defining constants and conversions 
pi = 3.141592
kb = 1.3807e-23 #kb in J/K
G = 6.674e-11 #G in m^3/(kg*s)
c = 2.9979e8 #c in m/s 
hbar = 1.6704e-36 #in J/s 
sigma = 5.6704e-34
JtoeV = 6.24e18 
EnergyToKg = 1.7827e-36 
sToYrs = 3.17e-8
t0 = ((M**3-M_rem**3)/3)*((16*pi**3*kb**4*G**2)/(sigma*hbar**4*c**6))*((1+ma.sqrt(1-a**2))/ma.sqrt(1-a**2))**4 #Lifetime in J/c^2*(1/kg)*s

t_eV = t0*JtoeV

t_lifes = t_eV*EnergyToKg
t_lt_yrs = t_lifes*sToYrs

print('Blackbody BH Lifetime is : {} years'.format(t_lt_yrs))
```
